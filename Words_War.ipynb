```python

!pip install deep-translator
import random
import nltk
from nltk.corpus import words
from deep_translator import GoogleTranslator

nltk.download('words')

web2lowerset = set(w.lower() for w in words.words())
```

```python
def translate_to_arabic(word):
    try:
        translation = GoogleTranslator(source='en', target='ar').translate(word)
        return translation
    except:
        return "Translation failed"


class Game():
    def __init__(self, word1, word2):
        self.word1 = word1.lower()
        self.word2 = word2.lower()
        self.is_valid = self.check_word()

    def check_word(self):
        if len(self.word1) > len(self.word2):
            print("Word 1 is longer than word 2")
            return False
        elif len(self.word1) < len(self.word2):
            print("Word 2 is longer than word 1")
            return False
        elif len(self.word1) == len(self.word2):
            if self.word1 in web2lowerset and self.word2 in web2lowerset:
                return True
            else:
                print("The words are not in the english dictionary")
                return False

    def start_game(self):
        if not self.is_valid:
            return

        word1_score = 0
        word2_score = 0
        letters_word1 = tuple(self.word1)
        letters_word2 = tuple(self.word2)

        for x in range(len(letters_word1)):
            if ord(letters_word1[x]) > ord(letters_word2[x]):
                word1_score += 1
                print(f"{letters_word1[x]} ({ord(letters_word1[x])}) won and {letters_word2[x]} ({ord(letters_word2[x])}) lost")
            elif ord(letters_word1[x]) < ord(letters_word2[x]):
                word2_score += 1
                print(f"{letters_word2[x]} ({ord(letters_word2[x])}) won and {letters_word1[x]} ({ord(letters_word1[x])}) lost")
            else:
                print(f"The letter {letters_word1[x]} ({ord(letters_word1[x])}) is equal to the letter {letters_word2[x]} ({ord(letters_word2[x])})")

        # إعلان الفائز
        if word1_score > word2_score:
            print("The winner is Word 1")
            print("\nMeanings & Translations:")
            print(f"The word of {self.word1} in Arabic is : {translate_to_arabic(self.word1)}")
            print(f"The word of {self.word2} in  Arabic is : {translate_to_arabic(self.word2)}")

        elif word1_score < word2_score:
            print("The winner is Word 2")
            print("\nMeanings & Translations:")
            print(f"The word of {self.word1} in Arabic is : {translate_to_arabic(self.word1)}")
            print(f"The word of {self.word2} in  Arabic is : {translate_to_arabic(self.word2)}")

        else:
            print("They are equal")

        # عرض المعاني والترجمة
            print("\nMeanings & Translations:")
            print(f"The word of {self.word1} in Arabic is : {translate_to_arabic(self.word1)}")
            print(f"The word of {self.word2} in  Arabic is : {translate_to_arabic(self.word2)}")


def AI_OR_PVP():
    print("Choose Game Mode:")
    print("1 - Player vs Player")
    print("2 - Player vs Computer")

    choice = input("Enter 1 or 2:")

    if choice == "1":
        word1 = input("Enter the first word: ")
        word2 = input("Enter the second word: ")
        return word1, word2

    elif choice == "2":
        word1 = input("Enter your word: ")
        same_length_words = [w for w in web2lowerset if len(w) == len(word1)]
        word2 = random.choice(same_length_words)
        print(f"Computer chose: {word2}")
        return word1, word2

    else:
      #Add like this to the check word
        print("Invalid choice, try again.")
        return AI_OR_PVP()


# ====== تشغيل اللعبة ======
word1, word2 = AI_OR_PVP()
game1 = Game(word1, word2)
game1.start_game()
```
